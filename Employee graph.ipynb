```python
import numpy as np
from datascience import *
import pandas as pd
import urllib.request, json
from datascience import Table
```

```python
pd.set_option("display.max_rows", 100)  # Show up to 100 rows
pd.set_option("display.max_columns", 50)
```

```python
URL = "https://script.google.com/macros/s/AKfycbyzumjN7OiWV1jOv1gAb0cjQkjAEgw8gL1JGYe7vb_GeT133HiLHTwhKMpkUEaMOl1r0A/exec"

# Fetch data from Google Apps Script
with urllib.request.urlopen(URL) as data_url:
    raw_data = json.load(data_url)
type(raw_data)
```

```python
assert raw_data['data']
```

```python
def rawToDf(raw):
    data = raw['data']
    arr = list(data.keys())
    output = []
    for i in arr:
        output.append(pd.DataFrame(data[i][1:], columns=data[i][0]))
    return output
```

```python
months = rawToDf(raw_data)
january = months[0]
january
```

```python
february = months[1]
february
```

```python
print(february['Total Sales'].tolist())

def makeTable(arr):
    total = Table()
    col1 = np.arange(1,len(arr)+1)
    col2 = []
    for i in arr:
        value = i['Total Sales'].tolist()
        value = value[len(value)-1]
        col2.append(value)
    total = total.with_column("Months", col1)
    total = total.with_column('Total Sales', col2)
    return total

total1 = makeTable(months)
total1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
total1.plot('Months','Total Sales')
```
